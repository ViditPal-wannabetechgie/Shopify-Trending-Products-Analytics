```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_sales = pd.read_csv('shopify_trending_products_2025.csv')

print(f"Datasetshape: {df_sales.shape}")
print("-"*50)
print("Columns in Dataset")
print(df_sales.columns.tolist())
print("-"*50)

df_sales.head()
```

```python
print("Missing values per column")
print("-"*50)
print(df_sales.isnull().sum())
print("-"*50)

print("/nData Types")
print(df_sales.info())
```

```python
category_revenue = df_sales.groupby('Category')['Estimated_Revenue_in_2025_USD'].sum().sort_values(ascending=False)

print("Top revenue generating categories(USD):")
print("*"*50)
print(category_revenue)

plt.figure(figsize=(10,6))

sns.barplot(x=category_revenue.values, y=category_revenue.index, palette='Blues_r')
plt.title("Top Revenue Generating Categories(USD)", fontsize=20, fontweight = "bold", pad = 15)
plt.xlabel("Total Revenue(USD)", fontsize = 15)
plt.ylabel("Category", fontsize = 15)
plt.show()
```

```python
source_revenue = df_sales.groupby('Trend_Source')['Estimated_Revenue_in_2025_USD'].sum()

plt.figure(figsize=(7,7))

plt.pie(source_revenue.values, labels=source_revenue.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'), pctdistance=0.85)

centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig=plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Which platform drives the most revenue?:", fontsize=14, fontweight='bold', pad=15)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data=df_sales, x="Trend_Score", y="Estimated_Revenue_in_2025_USD", color = 'dodgerblue', alpha=0.7, s=80, edgecolor='black', linewidth=0.5)
plt.title("Relationship between Trend Score and Estimated Revenue", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Trend Score", fontsize=12, fontweight='bold')
plt.ylabel("Estimated Revenue", fontsize=12, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
numeric_features = ['Estimated_Total_Units_Sold_in_2025', 'Estimated_Revenue_in_2025_USD', 'Trend_Score']
sales_corr  = df_sales[numeric_features].corr()

plt.figure(figsize=(10,6))
sns.heatmap(sales_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title("E_Commerce_sales : Feature correlation heatmap", fontsize=14, fontweight='bold', pad=15)
plt.show()
```
